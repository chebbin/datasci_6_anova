```r
df = read.csv('/content/diabetic_data.csv')
df
```

```r
## Anova test shows a very low P value which indicates that time in the hospital
## is affected by the race and A1C result

model <- aov(time_in_hospital ~ race * A1Cresult, data=df)

summary(model)
```

```r
means <- aggregate(time_in_hospital ~ race, data=df, FUN=mean)
print(means)
```

```r
means <- aggregate(time_in_hospital ~ A1Cresult, data=df, FUN=mean)
print(means)
```

```r
## When considering only race, P values are significant (<.05) when comparing
## African Americans and ?, African Americans and Asians, African Americans and
## Caucasians, African Americans and other. So every permutation with African
## Americans. Also, the P value is significant when comparing Hispanics and Caucasions.
## When comparing only AIC results, the ones that are insignificant are the ones
## that are close together, so when >8 and >7, as well as Normal and >7. This makes
## sense. The further the numbers are from each other, the more significant the
## difference will be. Normal and >8 is the example that shows this.
TukeyHSD(model)
```
